```python
import time
import speech_recognition as sr
import pyttsx3
from gtts import gTTS
import webbrowser

def recognize_speech():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)
        try:
            text = recognizer.recognize_google(audio)
            print(f"Recognized: {text}")
            return text
        except sr.UnknownValueError:
            print("Sorry, I did not understand that.")
        except sr.RequestError as e:
            print(f"Request error: {e}")
        return ""

def process_command(command):
    if "i am hungry" in command.lower():
        return "wanna have some Biryani today?"
    elif "time" in command.lower():
        from datetime import datetime
        return f"The current time is {datetime.now().strftime('%H:%M')}."
    elif "play music" in command.lower():
        url = "https://music.youtube.com/watch?v=hMACEmdz5Kk&si=D2EDiYXGxJN1fP-w"
        webbrowser.open_new_tab(url)
        return f"Playing music. Opening link in your web browser."
    elif "open website" in command.lower():
        url = "https://www.example.com"  # Replace with your desired URL
        webbrowser.open_new_tab(url)
        return f"Opening website: {url}"
    elif "goodbye" in command.lower():
        return "Goodbye! Have a nice day."
    elif "disha" in command.lower():
        return "Yes, how can I assist you?"
    else:
        return ""

def text_to_speech(text):
    engine = pyttsx3.init()

    # Female voice configuration
    voices = engine.getProperty('voices')
    engine.setProperty('voice', voices[1].id)  # Index 1 is typically female voice

    engine.say(text)
    engine.runAndWait()

# Define the AI's name
AI_NAME = "Disha"

# Greet the user initially
text_to_speech(f"Hello! I am {AI_NAME}. How can I assist you today?")

# Main interaction loop
while True:
    # Recognize speech
    recognized_text = recognize_speech()

    # Check for wake-up command
    if "disha" in recognized_text.lower():
        response = process_command(recognized_text)
        if response:
            print("Response:", response)
            # Speak the response with female voice
            text_to_speech(response)
            # Sleep for 3 seconds
            time.sleep(3)
    
    # Check for exit command
    if "good bye" in recognized_text.lower():
        break
    
    # Check for commands that open URLs in a new tab
    if "play music" in recognized_text.lower() or "open website" in recognized_text.lower():
        pass  # These commands will directly open the URLs in a new browser tab
```
